```python
import pandas as pd
import os  
import glob

df_global = pd.read_csv('csv_results/Categories.csv')
df_global.info()
df_global = df_global.drop(df_global.columns[[0]], axis=1, inplace=True)

df_compagnies = pd.read_csv('csv_results/Compagnies.csv')
df_compagnies.info()
df_compagnies = df_compagnies.drop(df_compagnies.columns[[0]], axis=1, inplace=True)

df_reviews = pd.DataFrame({'Reviewer': pd.Series(dtype='str'),
                          'Note': pd.Series(dtype='float'),
                          'Comment': pd.Series(dtype='str'),
                          'Localisation': pd.Series(dtype='str'),
                          'Nbreviews': pd.Series(dtype='int'),
                          'Date': pd.Series(dtype='str'),
                          'Compagny': pd.Series(dtype='str')})

path = "csv_results/*_Reviews.csv"
files = glob.glob(path)
print(files)

for file in files:
    df_review_cp = pd.read_csv(file)
    df_review_cp.info()
    df_review_cp = df_review_cp.drop(df_review_cp.columns[[0]], axis=1, inplace=True)
    #df_reviews = df_reviews.append(df_review_cp, ignore_index=True)

#df_reviews.head()
```

```python
#harmonize columns names 
df_reviews.rename(columns={"Compagny": "Compagny_Name"})
df_compagnies.rename(columns={"Nom": "Compagny_Name"})

# step 1 merge REVIEW + COMPAGNY
df_rev_com = pd.merge(df_compagnies, df_reviews, on = ['Compagny_Name'], how = 'outer')

# Step 2 merge Result of STEP 1 + CATEGORY
df_final = pd.merge(df_global, df_rev_com, on = ['SubCategory'], how = 'outer')

df_final.info()
```
